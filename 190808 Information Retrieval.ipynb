```python
import konlpy
```

```python
import sqlite3
con =sqlite3.connect('news.db')
cur = con.cursor()

cur.execute("""
    SELECT content
    FROM table2
    ORDER BY id ASC
""")

corpus = [_[0] for _ in cur.fetchall()]
```

```python
def eojeol(sentence, N =2):
    tokens = word_tokenize(sentence)
    ngram = list()
    for i in range(len(tokens)-(N-1)):
        ngram.append(tuple(tokens[i:i+N]))
    return ngram

def umjeol(token, N=2):
    tokens = list(token)   #음절로 쪼개짐.
    ngram = list()
    for i in range(len(tokens)-(N-1)):
        ngram.append("".join(tokens[i:i+N]))
    return ngram
```

# Document representation with BOW (Controlled vocab)

```python
corpus[0]
```

```python
import nltk
from string import punctuation
#word_tokenize(corpus[0])
#punktuation 날리려면 정규표현이 좀 필요할거다
import re
pattern1 = re.compile(r"[{0}]".format(re.escape(punctuation)))
pattern1.sub("",corpus[0])

pattern2 = re.compile(r"[A-Za-z]{5,}")
pattern2.sub("",pattern1.sub("",corpus[0]))

pattern3 = re.compile(r"[^ \n0-9A-Za-z가-힣]")
pattern3.sub("",pattern2.sub("",pattern1.sub("",corpus[0])))

pattern4 = re.compile(r"\s{2,}")
pattern4.sub(" ",
            pattern3.sub(" ",pattern2.sub(" ",pattern1.sub("",corpus[0]))))

collection = list()
for _ in corpus:
    collection.append(
        pattern4.sub(" ",
        pattern3.sub(" ",
        pattern2.sub(" ",
        pattern1.sub("",
        _))).strip())
)
```

```python
#Token화(어절, 형태소, Ngram(sequence저장위해))
#Ngram -> Stemming/Normalization(길이,빈도) -> Stopwords
from nltk.tokenize import word_tokenize
len(word_tokenize(collection[0])), len(collection[0].split()) #구두점을 다 날렸기때문에 똑같음
```

```python
from konlpy.tag import Kkma
kkma = Kkma()
kkma.pos('아직 목요일이라니..'), kkma.morphs('아직 목요일이라니..')
#단순 stemming이중요하면 morphs,,품사정보가 필요하면 pos 
#양쪽 정보(pos, morphs)가 모두필요할거다.. controlled vocab

#Kkma- 계산량 많으면 뻗음주의
```

```python
len(collection[0].split()),len(eojeol(collection[0])),\
len(umjeol(collection[0])),len(kkma.morphs(collection[0]))
```

```python
Vocabulary = list()
for _ in collection:
    Vocabulary.extend(_.split())
    Vocabulary.extend(eojeol(_))
    Vocabulary.extend(umjeol(_))
    Vocabulary.extend(kkma.morphs(_))
```

```python
len(Vocabulary), len(set(Vocabulary))  #unique한거 2만몇개.. /뉴스 35개가지고 이정도/
```

```python
# 지금까지 Ngram -> Stemming/Normalization(길이,빈도) -> Stopwords(비속어,어미,조사,감탄사,,,)
```

```python
#이제 문서 1개를 21392개의 차원으로 표현할거다.
```

```python
Vocabulary = [_ for _ in list(set(Vocabulary))
             if 1 < len(_) < 6]
len(Vocabulary) #단어의 길이가 2~5인애들.
```

```python
kkma.tagset
```

```python
#bag of word matrix (document를 term들로 표현한 dtm)
BoM = list(0 for _ in range(len(Vocabulary)))
for _ in collection[0].split() + eojeol(collection[0]) + umjeol(collection[0])+ kkma.morphs(collection[0]):
    if _ in Vocabulary:
        BoM[Vocabulary.index(_)] = 1
```

```python
len(Vocabulary)-sum(BoM), sum(BoM)  #0번 뉴스는 0이 18000개정도되는 sparse함.
```

```python
BoM = list()
for d in collection:
    docVector = list(0 for _ in range(len(Vocabulary)))
    for _ in d.split() + eojeol(d) + umjeol(d)+ kkma.morphs(d):
        if _ in Vocabulary:
            docVector[Vocabulary.index(_)] =1
    BoM.append(docVector)
```

#  여기까지 아래의 Doc Representation을 한것

```python
Vocabulary.index("국민"), Vocabulary.index("법안")
```

```python
Query = ['국민', '법안']
totalResult = []

for q in Query:
    searchResult = list()
    for i,d in enumerate(BoM):
        for j, b in enumerate(d):
            if b==1 and j==Vocabulary.index(q):
                searchResult.append(i)
                break
    totalResult.append(searchResult)
```
